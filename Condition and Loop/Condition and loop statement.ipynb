# Python if...elif...else Statement

```python
# if-elif statement example

letter = "A"

if letter == "B": 
    print("letter is B") # this body of if wont execute

elif letter == "C":
     print("letter is C") # this body of if wont execute

elif letter == "A":  # condition true
     print("letter is A") # this body of if will execute

else:
    print("letter isn't A, B or C")
```

```python
# you have accept the input from user and try to find the given number is +ve , -ve or zero
number = 0

if number > 0:
    print("Positive number")

elif number == 0:
    print('Zero')
else:
    print('Negative number')

print('This statement is always executed')
```

```python
number = 5

# outer if statement
if (number >= 0):
    # inner if statement
    if number == 0:
        print('Number is 0')
    
    # inner else statement
    else:
        print('Number is positive')

# outer else statement
else:
    print('Number is negative')

# Output: Number is positive
```
